```python
dnevne_spremembe = pd.to_numeric(delnice["Dnevna_sprem."].str.replace("%", ""), errors='coerce')
delnice["Dnevna_sprem_št"] = dnevne_spremembe
delnice[abs(dnevne_spremembe) > 5].sort_values("Dnevna_sprem_št", ascending= True)[["Ime", "Dnevna_sprem_št"]]
# To je v odstotkih, sam če imam znak "%" mi ne sortira pravilno
```
